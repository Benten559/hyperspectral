```python
import pandas as pd
import numpy
df_avgCoords = pd.read_csv("TripodAvgGPS1.csv")
df_coordsDT = pd.read_csv("Bens_WORKFILE_ST_07.csv")
# df_co.head() 
# df_avgCoords.join(df_CoordsDT,how ='inner',lsuffix =)
df_orderedCoords = df_avgCoords.set_index('img_id').join(df_coordsDT.set_index('Image')).sort_values('DT').reset_index()
df_orderedCoords['degree'] = ''
# df_avgCoords.head()
df_orderedCoords = df_orderedCoords[['img_id','latitude','longitude','degree']]
df_orderedCoords.head(3)
```

```python
numpy.arctan2(df_avgCoords.iloc[0,1] - df_avgCoords.iloc[1,1],df_avgCoords.iloc[0,2]-df_avgCoords.iloc[1,2])
```

```python
for i in range(len(df_avgCoords)-1):
    xDiff = df_orderedCoords.iloc[i,1] - df_orderedCoords.iloc[i+1,1]
    yDiff = df_orderedCoords.iloc[i,2]-df_orderedCoords.iloc[i+1,2]
    atan2 = numpy.arctan2(xDiff,yDiff)

    # print("Change in latitude" + str(xDiff))
    # print("Change in longitude "+str(yDiff))
    df_orderedCoords.iloc[i,3] = atan2
    print(atan2)
df_orderedCoords.to_csv('AllAvgCoordinates_Degrees.csv',index=False)

    
```
